# 数据聚合与分组运算

# GroupBy机制

```python
df = pd.DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'],
                  'key2': ['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2': np.random.randn(5)})
```

```python
df
```

```python
# 按key1进行分组，并计算data1列的平均值
grouped = df['data1'].groupby(df['key1'])
```

```python
grouped
```

```python
# 变量grouped是一个GroupBy对象，它实际上还没有进行任何计算，只是含有一些中间数据而已
# 调用GroupBy的mean方法计算分组平均值
grouped.mean()
```

```python
# 一次传入多个数组列表
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
```

```python
means
```

```python
# 通过两个键对数据进行了分组，得到的Series具有一个层次化索引
means.unstack()
```

```python
# 该例中分组键均为Series，实际上分组键可以是任何长度适当的数组
states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
```

```python
years = np.array([2005, 2005, 2006, 2005, 2006])
```

```python
df['data1'].groupby([states, years]).mean()
```

```python
# 还可以将列名（可以是字符串、数字或其他Python对象）用作分组键
df.groupby('key1').mean()
```

```python
df.groupby(['key1', 'key2']).mean()
```

```python
# size方法返回一个含有分组大小的Series
df.groupby(['key1', 'key2']).size()
```

## 对分组进行迭代

```python
# GrouopBy对象支持迭代，可以产生一组二元元组
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
# 对于多重键的情况，元组的第一个元素将会是又键值组成的元组
for(k1, k2), group in df.groupby(['key1', 'key2']):
    print((k1, k2))
    print(group)
```

```python
# 可以将这些数据片段做成一个字典
pieces = dict(list(df.groupby('key1')))
```

```python
pieces['b']
```

```python
# groupby默认在axis = 0上进行分组
# 可以根据dtype对列进行分组
df.dtypes
```

```python
grouped = df.groupby(df.dtypes, axis = 1)
```

```python
# 打印分组
for dtype, group in grouped:
    print(dtype)
    print(group)
```

## 选取一列或列的子集

```python
# 由DataFrame产生的GroupBy对象，如果用一个（单个字符串）或一组（字符串数据）列名对其进行索引
# 能实现选取部分列进行聚合
# 对于大数据集，很可能只需要对部分列进行聚合
# 如：计算data2列的平均值并以DataFrame形式得到结果
df.groupby(['key1', 'key2'])[['data2']].mean()
```

```python
# 这种索引操作所返回的对象是一个已分组的DataFrame（如果传入的是列表或数组）
# 或已分组的Series（如果传入的是标量形式的单个列名）
s_grouped = df.groupby(['key1', 'key2'])['data2']
```

```python
s_grouped
```

```python
s_grouped.mean()
```

```python
people = pd.DataFrame(np.random.randn(5, 5),
                     columns = ['a', 'b', 'c', 'd', 'e'],
                     index = ['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
```

```python
people.iloc[2:3, [1, 2]] = np.nan
```

```python
people
```

```python
# 假设已知列的关系，并希望根据分组计算列的和
mapping = {'a': 'red', 'b': 'red', 'c': 'blue',
          'd': 'blue', 'e': 'red', 'f': 'orange'}
```

```python
# 直接传递字典
by_column = people.groupby(mapping, axis = 1)
```

```python
by_column.sum()
```

```python
# Series可以被看做一个固定大小的映射
map_series = pd.Series(mapping)
```

```python
map_series
```

```python
people.groupby(map_series, axis = 1).count()
```
